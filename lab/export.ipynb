```python
import os
import re
```

```python
path = os.path.join(os.getcwd(), '..', 'src', 'stringdataframe')
```

```python
cmpl = re.compile('^proc (.+?)[*].*?')
for spe in os.walk(path):
    for file in spe[2]:
        print(f'----------{file}----------')
        with open(os.path.join(path, file), 'r', encoding='utf-8') as fp:
            for line in fp.readlines():
                m = cmpl.match(line)
                if m is not None:
                    print(f'export {m.groups()[0]}')
```
